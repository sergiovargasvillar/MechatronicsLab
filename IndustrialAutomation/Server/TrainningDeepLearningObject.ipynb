```
#Persistent Gdrive folder
import os
os.chdir("/content/gdrive/My Drive/TFM/keras-yolo3/")
#parse XML format example

from xml.dom import minidom

file = open("images/pieza_planta_piloto/images_2/train.txt","w")
for x in range(1,112):

  xmldoc = minidom.parse("images/pieza_planta_piloto/images_2/"+str(x)+'.xml')

  path = "images/pieza_planta_piloto/images_2/"+str(x)+".jpg "
  file.write(path)
  c = 0
  while True:
      try:
        xmin = xmldoc.getElementsByTagName("xmin")[c]
        ymin = xmldoc.getElementsByTagName("ymin")[c]
        xmax = xmldoc.getElementsByTagName("xmax")[c]
        ymax = xmldoc.getElementsByTagName("ymax")[c]
        obj =  xmldoc.getElementsByTagName("name")[c]
        if str(obj.firstChild.data) == "pieza":
         print (obj.firstChild.data)
         tobj = "0 ";
        elif str(obj.firstChild.data) == "triangulo":
         print (obj.firstChild.data)
         tobj = "1 "
      except IndexError:
          break
      else:
        value = str(xmin.firstChild.data) + ","+ str(ymin.firstChild.data) + ","+ str(xmax.firstChild.data) + ","+ str(ymax.firstChild.data) + ","+str(tobj)
        file.write(value)
        c = c+1
  file.write("\r")
  print(value)


file.close()
```
